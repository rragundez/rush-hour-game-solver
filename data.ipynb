```python
import matplotlib.pyplot as plt
from pandas import DataFrame
import matplotlib.pyplot as plt
from __future__ import division
%matplotlib inline
```

```python
time = [211.47,21.57,118.49,147.54]
moves = [18,53,45,60]
database = [2506,822,2015,2079]

df = DataFrame({'Execution Time':[t/max(time) for t in time], 'Number of Moves':[m/max(moves) for m in moves], 
                'Board Database Size':[d/max(database) for d in database]}, 
               index = ['Beginner','Intermediate','Advanced','Expert'])

ax = df.plot(marker='.',markersize=20,linewidth=3.0,fontsize = 20)
ax.set_ylabel("Normalized Quantities", fontsize = 20)
ax.legend(fontsize = 20, loc = 4)
plt.show()
```
